```python
''' 
Problem Statement: -
Kitabi Duniya, a famous book store in India, which was established before Independence,
the growth of the company was incremental year by year, but due to online selling of books and wide 
spread Internet access its annual growth started to collapse, seeing sharp downfalls, you as a Data Scientist help this 
heritage book store gain its popularity back and increase footfall of customers and provide ways the business can improve exponentially, 
apply Association RuleAlgorithm, explain the rules, and visualize the graphs for clear understanding of solution.
1.) Books.csv
'''
```

```python
# import the requerd librarise to load the data file
import pandas as pd
df = pd.read_csv('book.csv')
df.head()
```

```python
# Data Preprocessing 
# check the null values
df.isnull().sum()
```

```python
# drop duplicates if any 
df.drop_duplicates(inplace=True)
```

```python
# Standerize the data 
from sklearn.preprocessing import StandardScaler
sd = StandardScaler()
sd.fit_transform(df)
```

```python
# convert the dataset into list of lists
# why we need to convert it into list formate
# 1. transaction data format: 
    # Apriori algorithm only work on Trasactional data where each transaction is collection of items.
    # eg. ['name','class']
# 2. Apriori algorothm input 
    # Input should be in the list format
# 3. Transaction Encoding
    # when the data is in the list format then mlxtend can easily apply transactionencoders
    # to convert the data into a formate suitable for the Apriori algorithm
    # TransactionEncoder transform the list of lists into a "one-hot-encoded" Dataframe
    # each column represents an item, and each row corresponds to a transaction.
    # the value is 1 if an item is present in the transaction 
    # 0 if it is absent.
# process
# initilize the empty list 
tran = []
for i in range(0, len(df)):
    tran.append([str(df.values[i,j]) for j in range(0, len(df.columns)) if pd.notna(df.values[i,j])])
    

print(tran)   
```

# Model building

```python
# the Apriori algorithm identifies frequent itemsets and then uses these itemsets to generate association rules.
# use the apriori method from the mlxtend library to mine the frequnt items
# import the requierd librarise for Apriori 
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

# step 1 convert the data set into a formate suitable for apriori
te = TransactionEncoder()
te_arr = te.fit(tran).transform(tran)
df1 = pd.DataFrame(te_arr,columns=te.columns_)
```

```python

# step 2 : apply the apriori algorith to find the frequent itemset 
fre_item = apriori(df, min_support=0.1, use_colnames=True)
```

```python
# step 3 : generate association rules from frequent itemset 
rule = association_rules(fre_item, metric='lift', min_threshold=1)
```

```python
# output of the result 
print("Freuentnt itemset\n ")
print(fre_item)
```

```python
print("\n association rule \n")
print(rule[['antecedents','consequents','support','confidence','lift']])
```

```python
''' 
Observation:

The rule (ChildBks) -> (YouthBks) with a confidence of 0.49 and a lift of 1.08 shows that customers who buy children's books also tend to buy youth books, though the relationship is moderately strong.
Similarly, (YouthBks) -> (ChildBks) with a higher confidence of 0.67 and a lift of 1.08 suggests a stronger tendency for customers who purchase youth books to also buy children's books.

This data-driven approach can help Kitabi Duniya regain its competitive edge in the market.
'''
```

```python
''' 
Objective:1
The primary objective of this analysis is to help Kitabi Duniya, a heritage bookstore in India,
regain its market position and increase customer footfall. By applying the Association Rule Algorithm, 
the aim is to uncover meaningful relationships between different book categories that customers frequently purchase together. 
'''
```
